```python
stat = pd.read_csv('data/General Statistics For each Participated Team.csv')
ranking = pd.read_csv('data/fifa_ranking-2023-07-20.csv')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.britannica.com/sports/Africa-Cup-of-Nations'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')

years = []
winners = []
runners_up = []

for row in table.find_all('tr')[1:]:  
    cells = row.find_all('td')
        
    if len(cells) >= 3:  
        years.append(cells[0].get_text(strip=True))
        winners.append(cells[1].get_text(strip=True))
        runners_up.append(cells[2].get_text(strip=True))

champions = pd.DataFrame({
    'year': years,
    'winner': winners,
    'runner-up': runners_up
    })
```

## Data processing

```python
ranking = ranking[ranking['confederation'] == 'CAF']
ranking = ranking.drop('confederation',axis = 1)
ranking.reset_index(drop=True, inplace=True)
```

```python
ranking.drop(['country_abrv', 'total_points', 'previous_points'], axis=1, inplace=True)
```

```python
ranking['year'] = ranking['rank_date'].str.split('-').str[0].astype(int)
ranking.drop(['rank_date'], axis=1, inplace=True)
ranking.rename(columns={'country_full': 'team'}, inplace=True)
```

```python
ranking.head()
```

```python
champions['year'] = champions['year'].str.replace(r'\D', '', regex=True)
champions['year'] = champions['year'].astype(int)
```

```python
data = ranking.copy()
```

```python
data['year'] = data['year'].astype(int)
champions['year'] = champions['year'].astype(int)

champions = champions.sort_values('year')

data['previous_cup_wins'] = data.apply(lambda row: len(champions[(champions['year'] <= row['year']) & (champions['winner'] == row['team'])]), axis=1)
data['won_this_year'] = data.apply(lambda row: 1 if ((champions['year'] == row['year']) & (champions['winner'] == row['team'])).any() else 0, axis=1)
```

```python
champions['winner'] = champions['winner'].replace({'Congo (Kinshasa)': 'DR Congo'})
champions['winner'] = champions['winner'].replace({'Congo (Brazzaville)': 'Congo'})
```

```python
mapping_dict = {
    'Cabo Verde': 'Cape Verde',
    'Cape Verde Islands': 'Cape Verde',
    'Central African Republic': 'CAR',
    'Congo DR': 'DR Congo',
    "Côte d'Ivoire": 'Ivory Coast',
    'Sao Tome e Principe': 'Sao Tome and Principe',
    'São Tomé e Príncipe': 'Sao Tome and Principe',
    'Swaziland': 'Eswatini',
    'São Tomé and Príncipe': 'Sao Tome and Principe',
    'The Gambia': 'Gambia',
    'Zaire': 'DR Congo'
}

data['team'] = data['team'].replace(mapping_dict)
```

```python
data = data.merge(stat[['Team', 'Part', 'W', 'D', 'L', 'Gd', 'Pts']], left_on='team', right_on='Team', how='left')
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data = data.drop('Team',axis=1)
```

```python
data
```

```python
qualified_teams_afcon2023 = [
    "Nigeria",
    "Cameroon",
    "Angola",
    "Namibia",
    "Cape Verde", 
    "Mauritania",  
    "Morocco",
    "South Africa",  
    "Mali",
    "Burkina Faso",
    "Senegal",
    "Ivory Coast",  
    "Egypt",
    "DR Congo",  
    "Equatorial Guinea",  
    "Guinea"  
]
```

```python
data_2023 = data[data['year'] == 2023]

teams_2023 = set(data_2023['team'])

all_teams_present = all(team in teams_2023 for team in qualified_teams_afcon2023)

print(all_teams_present)
```

```python
filtered_data = data[(data['year'] == 2023) & (data['team'].isin(qualified_teams_afcon2023))]
```

```python
test = filtered_data[-16:]
test.reset_index(drop=True, inplace=True)
```

```python
train = data[data['year'] != 2023]
train.reset_index(drop=True, inplace=True)
```

## Model

```python
from xgboost import XGBClassifier

model = XGBClassifier(learning_rate = 0.01,random_state=42)

X_train = train.drop(['team', 'won_this_year'], axis=1)
y_train = train['won_this_year']

X_test = test.drop(['team', 'won_this_year'], axis=1)

model.fit(X_train, y_train)

y_probabilities = model.predict_proba(X_test)[:, 1]

y_pred = model.predict(X_test)
```

```python
prediction = {'team':test['team'],'win proba':y_probabilities}
```

```python
prediction_df = pd.DataFrame(prediction)
```
